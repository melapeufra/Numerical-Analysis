```python
def poly(x,a,b,c):
    return(x**3+a*x**2+b*x+c)
def dp(x,a,b,c):
    return(3*x**2+a*2*x+b)
def newton(a,b,c,x0,niter,tol):
    nit=0
    res=np.zeros(niter)
    res[0]=abs(poly(x0,a,b,c))
    x=x0
    while(nit<niter) and (res[nit]>tol):
        nit +=1
        x=x-poly(x,a,b,c)/dp(x,a,b,c)
        res[nit]=abs(poly(x,a,b,c))
    return(x,nit,res[: nit+1])




```

```python
def polyN(x,coef):
    N =len(coef)
    y=x**(N+1)
    for i in range(N):
        y+=x**i*coef[i]
    return y
def dpolyN(x,coef):
    N=len(coef)
    y=(N+1)*x**(N-1)
    for i in range(N):
        y+=i*x**(i-1)*coef[i]
    return y  
    
def G_newton(coef,x0,niter,tol):
    nit=0
    res=np.zeros(niter+1)
    res[0]=abs(polyN(x0,coef))
    x=x0
    while(nit<niter) and (res[nit]>tol):
        nit +=1
        x=x-polyN(x,coef)/dpolyN(x,coef)
        res[nit]=abs(polyN(x,coef))
    return(x,nit,res[: nit+1])
coef=[0,-1,0]
print(G_newton(coef,1/3,100,1e-7)[2])
plt.plot(G_newton(coef,1/3,100,1e-7)[2])
```
